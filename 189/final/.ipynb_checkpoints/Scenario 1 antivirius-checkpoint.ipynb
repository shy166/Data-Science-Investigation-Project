```python
def analysis(col):
    temp = train[col].value_counts()
    temp = temp[temp > temp.iloc[0]//100].index
    #temp = train.index
    temp_df = train[train[col].isin(temp)]
    #temp_df = temp_df.sample(10000)
    p = temp_df.HasDetections.mean()
    q = 1-p
    random_data = pd.Series(np.random.choice([1,0], len(temp_df),p=[p,q]), index = temp_df.index).rename('HasDetections')
    temp_df['random_data'] = random_data
    
    #temp_df.groupby(col).HasDetections.mean().plot(kind = 'bar', figsize=(20,10))
    temp_df.groupby(col)[['HasDetections', 'random_data']].mean().plot(kind = 'bar', figsize=(20,10))
    plt.title('Percent of Has detections by {}'.format(col))
    
    display(pd.DataFrame({'train_data': temp_df.groupby(col).HasDetections.mean(),
                         'random_data': temp_df.groupby(col).random_data.mean()}))
    return stats.stats.ks_2samp([temp_df.groupby(col).HasDetections.mean(),
                temp_df.groupby(col).random_data.mean()])
```

```python
def permutation_test(col, n, trial):
    idx = train[col].value_counts().iloc[:n].index
    temp = train[train[col].isin(idx)]
    N = len(temp)
    def total_variation_distance(dist1, dist2):
        return np.sum(np.abs(dist1 - dist2)) / 2
    original_T = temp[temp.HasDetections == 1][col].value_counts(normalize = True, dropna = False)
    original_F = temp[temp.HasDetections == 0][col].value_counts(normalize = True, dropna = False)
    tvd_o = total_variation_distance(original_T, original_F)
    tvds = []
    for i in range(trial):
        temp[col] = temp[col].sample(replace=False, n=N).reset_index(drop=True)
        sample_T = temp[temp.HasDetections == 1][col].value_counts(normalize = True, dropna = False)
        sample_F = temp[temp.HasDetections == 0][col].value_counts(normalize = True, dropna = False)
        tvds += [total_variation_distance(sample_T, sample_F)]
    print(tvds, tvd_o)
    tvds = (np.array(tvds) >= tvd_o.mean()).mean()
    return tvds
    
```
